# Regression

### Importing the Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
pip install pycaret
from pycaret.datasets import get_data
from pycaret.regression import *
mpl.rcParams['figure.dpi'] = 300
```

### Loading the Dataset

```python
data = get_data('insurance')
```

```python
data.info()
```

### Exploratory Data Analysis
Exploratory Data Analysis (EDA) is the process of analyzing and understanding the characteristics of a dataset.

```python
# Create a list of numeric columns
numeric = ['age', 'bmi', 'children', 'charges']

# Plot histograms for numeric columns with 20 bins and a figure size of 12x8 inches
data[numeric].hist(bins=20, figsize=(12, 8))

# Display the plot
plt.show()
```

```python
# Create a list of categorical columns
categorical = ['smoker', 'sex', 'region']

# Create a list of colors for the bars in the plots
color = ['C0', 'C1', 'C2', 'C3']

# Create a figure with 2 rows and 2 columns, and a size of 12x10 inches
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Turn off the axis for the bottom right plot
axes[1, 1].set_axis_off()

# Loop through each axis and categorical column, and create a bar plot
for ax, col in zip(axes.flatten(), categorical):
    data[col].value_counts().plot(kind='bar', ax=ax, color=color)
    ax.set_xlabel(col)
```

```python
# Create a 2x2 subplot figure with a size of 12x8 inches
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Turn off the axis for the bottom right plot
axes[1, 1].set_axis_off()

# Loop through each axis and categorical column, and create a stacked histogram using seaborn's histplot()
for ax, col in zip(axes.flatten(), categorical):
    sns.histplot(data, x='charges', hue=col, multiple='stack', ax=ax)
```

```python
# Create a list of columns to be included in the pairplot
cols = ['age', 'bmi', 'charges', 'smoker']

# Create a pairplot with the specified columns and hue set to 'smoker'
sns.pairplot(data[cols], hue='smoker')

# Display the plot
plt.show()
```

### Initializing the PyCaret Environment

```python
reg = setup(data=data, target='charges', train_size = 0.8, session_id = 7402,
            numeric_features = numeric[:-1], categorical_features = categorical,
            transformation = True, normalize = True)
```

```python
# Retrieve the transformed feature matrix from the PyCaret pipeline and display the first few rows
get_config('X_transformed').head()
```

### Comparing Regression Models

```python
best = compare_models(sort='RMSE')
```

### Creating the Model

```python
model = create_model('gbr', fold = 10)
```

### Tuning the Model

```python
# Define a dictionary of hyperparameters to be tuned
params = {
    'learning_rate': [0.05, 0.08, 0.1],
    'max_depth': [1, 2, 3, 4, 5, 6, 7, 8],
    'subsample': [0.8, 0.9, 1, 1.1],
    'n_estimators': [100, 200, 300, 400, 500]
}

# Tune the specified model using the specified hyperparameters
tuned_model = tune_model(model, optimize='RMSE', fold=10, custom_grid=params, n_iter=100)
```

### Making Predictions

```python
# Generate predictions on the training data using the trained model
predictions = predict_model(model)
predictions.head()
```

### Plotting the Model



*   A **feature importance plot** is a graphical representation of the importance of each feature (or predictor) in a machine learning model.
*   It shows the contribution of each feature towards the final prediction made by the model. 
*    By visualizing the feature importances, we can gain insights into which features are most influential in predicting the target variable.







```python
# Plot the feature importances of the tuned model using the 'feature' parameter
plot_model(tuned_model, 'feature', scale = 4)
```

```python
# Plot the model errors using the 'error' parameter
plot_model(model, 'error')
```

### Finalizing and Saving the Model

```python
# Finalize the tuned model and save it for future use
final_model = finalize_model(tuned_model)

# Save the model to disk
save_model(final_model, 'regression_model')
```
